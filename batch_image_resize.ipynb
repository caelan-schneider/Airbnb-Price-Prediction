```python
from PIL import Image
from torchvision import transforms
import pandas as pd
import numpy as np

#ran this for each city CSV file
data = pd.read_csv("./data/la_listings.csv")
data["filename"] = data.picture_url.apply(lambda k: "images/"+k.split("/")[-1])
data = data[data.duplicated("picture_url", keep="first") == False]
data
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224)
])

for i in range(data.shape[0]):
    filename = "./" + data.iloc[i].filename
    try:
        img = Image.open(filename)
        img = preprocess(img)
        img.save(filename)
    except:
        data.iloc[i].filename = np.nan
```
